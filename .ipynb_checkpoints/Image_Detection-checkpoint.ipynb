```python
import cv2
import matplotlib.pyplot as plt
import urllib.request
```

```python
img = cv2.imread("img/OIP.jfif")
eyes = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Detect eyes
detected_eyes = eyes.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
# Draw rectangles
for (x, y, w, h) in detected_eyes:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
# Show result
# Convert BGR → RGB for matplotlib
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Display image
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Eye Detection")
plt.show()
```

```python
img = cv2.imread("img/OIP.jfif")
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Detect faces first
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
# Detect eyes
detected_eyes = eyes.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)
# Draw rectangles
for (x, y, w, h) in detected_eyes:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)
for (x, y, w, h) in faces:
    # Draw face rectangle
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = gray[y:y + h, x:x + w]
    roi_color = img[y:y + h, x:x + w]
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Face Detection")
plt.show()
```

```python
img = cv2.imread("img/OIP.jfif")

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
for (x, y, w, h) in detected_eyes:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0,255,0), 2)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = gray[y:y + h, x:x + w]
    smiles = smile_cascade.detectMultiScale(roi_gray,scaleFactor=1.3,minNeighbors=15,minSize=(20, 20))
    for (sx, sy, sw, sh) in smiles:
      cv2.rectangle(img, (x + sx, y + sy),(x + sx + sw, y + sy + sh), (0, 255, 0), 2)

# Convert BGR → RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Face, Eye and Smile Detection")
plt.show()
```

```python
img = cv2.imread("img/2.jpg")

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.3, minNeighbors=5)
count = 0 

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = gray[y:y + h, x:x + w]
    smiles = smile_cascade.detectMultiScale(roi_gray,scaleFactor=1.3,minNeighbors=8,minSize=(15, 15))
    for (sx, sy, sw, sh) in smiles:
      cv2.rectangle(img, (x + sx, y + sy),(x + sx + sw, y + sy + sh), (0, 255, 0), 2)
    count +=1 
# Convert BGR → RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.title("Grouped Smile Detection")
plt.show() 
print(f'{count} number of faces detected') 
```

### Face Detection using CV2 

```python
img = cv2.imread("img/girl-1.png")
pad = 40
img_padded  = cv2.copyMakeBorder(img, pad, pad, pad, pad, cv2.BORDER_REPLICATE)
gray_padded = cv2.cvtColor(img_padded, cv2.COLOR_BGR2GRAY)

eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye_tree_eyeglasses.xml")
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")
```

```python
faces = face_cascade.detectMultiScale(gray_padded, scaleFactor=1.03, minNeighbors=4,minSize=(200,200))


for (x, y, w, h) in faces:

    cv2.rectangle(img_padded, (x+pad, y+pad), (x+w-pad, y+h), (0, 255, 0), 3)
    roi_gray  = gray_padded[y:y+h, x:x+w]
    roi_color = img_padded[y:y+h, x:x+w]

    roi_gray_upper  = roi_gray[:int(roi_gray.shape[0] * 0.55), :]
    roi_color_upper = roi_color[:int(roi_color.shape[0] * 0.55), :]

    eyes = eye_cascade.detectMultiScale(roi_gray_upper,scaleFactor=1.1,minNeighbors=3,minSize=(15, 15),maxSize=(80, 80))
    print(f"Eyes detected: {len(eyes)}")
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color_upper, (ex, ey), (ex+ew, ey+eh), (255, 0, 255), 2)

    roi_gray_lower  = roi_gray[roi_gray.shape[0]//2:, :]
    roi_color_lower = roi_color[roi_color.shape[0]//2:, :]

    smiles = smile_cascade.detectMultiScale(roi_gray_lower,scaleFactor=1.3,minNeighbors=20,minSize=(25, 25))
    print(f"Smiles detected: {len(smiles)}")
    for (sx, sy, sw, sh) in smiles:
        cv2.rectangle(roi_color_lower, (sx, sy), (sx+sw, sy+sh), (255, 255, 0), 3)


img_rgb = cv2.cvtColor(img_padded, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Face (green), Eyes (magenta), Smile (cyan)")
plt.show()
```

```python
face_net     = cv2.dnn.readNetFromCaffe("deploy.prototxt", "res10_300x300_ssd_iter_140000.caffemodel")
eye_cascade  = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
smile_cascade= cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_smile.xml")

img = cv2.imread("img/girl-1.png")
pad  = 80
img_padded  = cv2.copyMakeBorder(img,  pad, pad, pad, pad, cv2.BORDER_REPLICATE)
gray_padded = cv2.cvtColor(img_padded, cv2.COLOR_BGR2GRAY)
gray_padded = cv2.equalizeHist(gray_padded)

h, w = img_padded.shape[:2]
print(f"Padded image shape: {img_padded.shape}")


blob = cv2.dnn.blobFromImage(img_padded, 1.0, (300, 300), (104, 117, 123))
face_net.setInput(blob)
detections = face_net.forward()

faces_found = 0

for i in range(detections.shape[2]):
    confidence = detections[0, 0, i, 2]
    if confidence > 0.5:
        faces_found += 1
        box = detections[0, 0, i, 3:7] * [w, h, w, h]
        x1, y1, x2, y2 = box.astype(int)

    
    
        cv2.rectangle(img_padded, (x1, y1), (x2, y2), (0, 255, 0), 3)
        
        roi_gray  = gray_padded[y1:y2, x1:x2]
        roi_color = img_padded[y1:y2, x1:x2]

        roi_gray_upper = roi_gray[:roi_gray.shape[0]//2, :]
        roi_color_upper = roi_color[:roi_color.shape[0]//2, :]
        
        eyes = eye_cascade.detectMultiScale(roi_gray_upper,scaleFactor=1.1,minNeighbors=5,minSize=(20, 20))
        print(f"Eyes detected: {len(eyes)}")
        
        for (ex, ey, ew, eh) in eyes:
            cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (255, 0, 255), 2)
    
            # ─── 6. Smile detection (lower half of face ROI) ──────────────────
        roi_gray_lower = roi_gray[roi_gray.shape[0]//2:, :]
        smiles = smile_cascade.detectMultiScale(roi_gray_lower,scaleFactor=1.3,minNeighbors=20,minSize=(25, 25))
        print(f"Smiles detected: {len(smiles)}")
        for (sx, sy, sw, sh) in smiles:
            sy += roi_gray.shape[0] // 2  # shift back to full face coords
            cv2.rectangle(roi_color, (sx, sy), (sx+sw, sy+sh), (255, 255, 0), 3)


print(f"Faces detected: {faces_found}")

# ─── 7. Display & save ────────────────────────────────────────────────────────
img_rgb = cv2.cvtColor(img_padded, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title("Face (green), Eyes (magenta), Smile (cyan)")
plt.show()

cv2.imwrite("final_detection.jpg", img_padded)
print("Saved final_detection.jpg")
```
